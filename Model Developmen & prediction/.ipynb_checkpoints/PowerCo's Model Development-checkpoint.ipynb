# Modelling

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df=pd.read_csv("churn_data.csv")
```

```python
df.head()
```

```python
df.info()
```

## Split the data

```python
# Separate target variable from independent variables
X = df.drop(columns=['id', 'churn'])
y = df['churn']
```

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

print ('Training cases: %d\nTest cases: %d' % (X_train.shape[0], X_test.shape[0]))
```

## Rectify the class Imbalancs

```python
from imblearn.over_sampling import SMOTE
X, y = SMOTE().fit_resample(X, y)
```

```python

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Modeling

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Precision, Recall, F1
precision = precision_score(y_test, y_pred, average='macro')  # use 'binary' for binary classification
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Visualize
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid", font_scale=1.2)  # Clean background and proper font scaling

sns.heatmap(
    cm, 
    annot=True,  # Annotate with actual counts
    fmt='d',     # Display as integers
    cmap='Blues',  # Color palette
    cbar=False,  # Disable color bar for simplicity
    annot_kws={"size": 14, 'weight': 'bold'},  # Font properties for annotations
    linewidths=0.5,  # Thin lines for grid separation
    linecolor='gray'  # Gridline color
)

plt.title('Confusion Matrix', fontsize=18, weight='bold')
plt.xlabel('Predicted Labels', fontsize=14, weight='bold')
plt.ylabel('True Labels', fontsize=14, weight='bold')

plt.xticks(ticks=[0.5, 1.5], labels=['Class 0', 'Class 1'], fontsize=12, rotation=0)
plt.yticks(ticks=[0.5, 1.5], labels=['Class 0', 'Class 1'], fontsize=12, rotation=0)

plt.tight_layout()

plt.show()
```

```python
# Full report
report = classification_report(y_test, y_pred)

print("Classification Report:\n", report)
```

## Prediction Comparison

```python
comparison_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})
```

```python
comparison_df.sample(20)
```

### Feature Importances

```python
feat_imp_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=True).reset_index()
```

```python
plt.figure(figsize=(10, 8))
sns.set(style="whitegrid")  # Clean background
sns.barplot(
    x="Importance",
    y="Feature",
    data=feat_imp_df,
    palette="viridis"
)

plt.title("Top 20 Feature Importances", fontsize=16, weight='bold')
plt.xlabel("Relative Importance", fontsize=12)
plt.ylabel("Feature", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
